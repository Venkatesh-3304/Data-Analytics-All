# 1. Write a Python program that takes a year as input from the user and prints whether it is a leap year or not.


```python
year = int(input("Enter a Year: "))
if (year % 4 == 0 and year % 100 != 0)  or (year % 400 == 0):
  print(f"{year} is a Leap Year")
else:
  print(f"{year} is NOT a Leap Year")
```

# 11. Write a Python program to extract the middle element(s) from a list or tuple using slicing.


```python
e = [2, 5, 7, 6, 5, 4]

mid = len(e) // 2

if len(e) % 2 == 0:
    middle = e[mid-1 : mid+1]
else:
    middle = e[mid : mid+1]

print(f"Middle element(s) = {middle}")
```

# 12. Write a Python program to concatenate two lists and print the result.

```python
family = ['Mani', 'Sankar', 'Saraswathi', 'Lakshmana Rao', 'Maggi', 'Mokshi']
frds = ['Ajay', 'Akash', 'Mohan', 'Gopi', 'Lali', 'Harika']

# method 1

fam_frd = family + frds
print(fam_frd)

# method 2

family.extend(frds)
print(fam_frd)

# method 3

for i in frds:
  family.append(i)
print(family)
```

# 13. Write a Python program to find the intersection of two lists and print the result.


```python
list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]

# Method 1

result = []

for i in list1:
  if i in list2:
    result.append(i)
print(f"Intersection = {result}")

# Method 2

result = list(set(list1) & set(list2))
print("Intersection", result)
```

# 14. Write a Python program to find the union of two lists and print the result.



```python
list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]
result = []

for i in list1:
    result.append(i)

for i in list2:
    if i not in result:
        result.append(i)

print(f"Union = {result}")

# method 2

result = list(set(list1) | set(list2))
print(result)  # [1, 2, 3, 4, 5, 6, 7]
```
