# 1. Set Operations - **Common Elements Finder**: Write a function that takes two lists as input and returns a set containing the common elements between the two lists.


```python
# Method 1


l1 = [2,4,6,8,10,1,3]
l2 = [3,5,3,6,4,11]

def common_elements(l1, l2):
  return set(l1) & set(l2)

print(common_elements(l1, l2))

# Method 2 (Using .intersection())

def common_elements(l1, l2):
  return set(l1).intersection(set(l2))
print(common_elements(l1, l2))


```

# 3. Function Practice: Write a program that defines a function to calculate the **factorial** of a given number and then calls this function to calculate the factorial of 5.


```python
def factorial(num):
  result = 1
  for i in range(1, num+1):
    result = result * i
  return result

print(factorial(5))
```

# 6. Function Practice - **Fibonacci Sequence**: Write a function to generate the Fibonacci sequence up to a specified number of terms.


```python
def fibonacci(n):
  fib = [0, 1]
  for i in range(2, n):
    next_val = fib[-1] + fib[-2]
    fib.append(next_val)
  return fib
print(fibonacci(8))
```

# 7. Function Practice - **Prime Number Checker**: Write a function that takes an integer as input and returns True if it's a prime number, and False otherwise.


```python
def is_prime(n):
  if n < 2:
    return False
  for i in range(2, n):
    if n % i == 0:
      return False
  return True

print(is_prime(8))


```


# 8. Function Practice - **Power Calculator**: Write a function that takes two numbers, base and exponent, as input and calculates the result of base raised to the power of exponent.


```python
def power_calculator(base,exponent):
  result = 1
  for i in range(exponent):
    result = result * base
  return result

print(power_calculator(2,3))
```

# 9. Function Practice - **Average Calculator**: Write a function that takes a list of numbers as input and returns the average of those numbers.


```python
l = [2, 5, 6, 7, 4, 5, 9]

def avg_calculator(l):
  if len(l) == 0:
    return 0

  return sum(l) / len(l)

print(avg_calculator(l))
```

# 10. Function Practice - **List Element Squaring**: Write a function that takes a list of numbers as input and returns a new list containing the square of each element.


```python
# Method 1

def square_list(numbers):
  new_list = []
  for num in numbers:
    new_list.append(num ** 2)
  return new_list

print(square_list([1, 2, 4, 5, 6, 7, 3]))


#Method 2

def square_list(numbers):
  return[num ** 2 for num in numbers]
print(square_list([1, 2, 4, 5, 6, 7, 3]))
```

# 17. Function Practice - **List Average (Exclude Zeros):** Write a function that takes a list of numbers as input and returns the average of the non-zero elements in the list.


```python
# write your code here
l = [2,4,6,8,0,5,0]

def ListAverage(l):
  final_list = [elem for elem in l if elem != 0]
  return sum(final_list) / len(final_list)

ListAverage(l)
```

# 18. Function Practice - **List Element Product**: Write a function that takes a list of numbers as input and returns the product of all the elements in the list.


```python
# write your code here
l = [2,4,6,8,5]

def listProduct(l):
  res = 1
  for elem in l:
    res *= elem

  return res

listProduct(l)
```

# 19. Function Practice - **Prime Factorization**: Write a function that takes an integer as input and returns a list of its prime factors.


```python
# write your code here
# 10 : 1, 2, 5, 10
# 16: 1, 2, 4, 8, 16

def Factorization1(num):
  res = []
  for i in range(1, num+1):  # n iterations  # O(n)
    if(num % i == 0):
      res.append(i)
  return res

# optimized code
def Factorization2(num):
  res = []
  for i in range(1, int(num/2) + 1):
    if(num % i == 0):
      res.append(i)

  res.append(num)
  return res

Factorization2(16)
```
