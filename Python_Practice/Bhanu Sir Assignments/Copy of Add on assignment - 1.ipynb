### 10. Write a Python program to slice a given string into three equal parts.


```python
# write your code here

s = "sdfkefdifjsf"
print("First Part", s[:len(s)//3])
print("Second Part", s[len(s)//3:2*len(s)//3])
print("Third Part", s[2*len(s)//3:])

print("\n")

s = [1,2,3,4,5,6]
print("First Part", s[:len(s)//3])
print("Second Part", s[len(s)//3:2*len(s)//3])
print("Third Part", s[2*len(s)//3:])

```
