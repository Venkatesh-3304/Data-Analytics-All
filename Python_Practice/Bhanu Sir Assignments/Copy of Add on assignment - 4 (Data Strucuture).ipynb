# 1. List Sum: Write a program that takes a list of numbers as input and calculates the sum of all the numbers in the list.


```python
numbers = list(map(float, input("Enter numbers seperated by spaces: ").split()))
total = sum(numbers)
print(f"Sum = {total}")
```

# 2. List Reversal: Write a program that takes a list of numbers as input and reverses the order of elements in the list.


```python
numbers = list(map(int, input("Enter numbers seperated by spaces: ").split()))
reverse = numbers[::-1]
print(f"Original : {numbers}")
print(f"Reversed : {reverse}")
```

# 3. List Sorting: Write a program that takes a list of numbers as input and sorts the list in ascending order.


```python
# Method 1

numbers = list(map(int, input("Enter numbers seperated by spacess: ").split()))
numbers.sort
print(numbers)

# Method 2

numbers = list(map(int, input("Enter numbers seperated by spaces: ").split()))
ascending_order = sorted(numbers)
print(f"Original : {numbers}")
print(f"Ascending : {ascending_order}")
```

# 4. List Filtering: Write a program that takes a list of numbers as input and filters out the even numbers, storing them in a new list.


```python
numbers = list(map(int, input("Enter numbers seperated by spaces: ").split()))
even_numbers = [num for num in numbers if num % 2 == 0]
print(f"Original : {numbers}")
print(f"Even Numbers : {even_numbers}")
```

# 5. List Element Search: Write a program that takes a list of numbers and a target number as input and checks if the target number is present in the list.


```python
numbers = list(map(int, input("Enter numbers seperated by spaces: ").split()))
target = int(input("Enter target number to search"))

if target in numbers:
  print(f"{target} i present in the list at index {numbers.index(target)}")
else:
  print(f"{target} is Not Present in the list")
```

# 6. Dictionary Key Access: Write a program that iterates over the keys of a dictionary and prints each key.


```python
person = {"name": "Venkatesh", "age": 22, "city": "vizag"}
for keys in person:
  print(keys)
```

# 7. Dictionary Value Access: Write a program that iterates over the values of a dictionary and prints each value.


```python
my_dict = {
    "A": 10,
    "B": 20,
    "C": 30
}

for value in my_dict.values():
  print(value)
```

# 8. Dictionary Item Access: Write a program that iterates over the items (key-value pairs) of a dictionary and prints each item.


```python
my_dict = {
    "A": 10,
    "B": 20,
    "C": 30
}

for pairs in my_dict.items():
  print(pairs)
```

# 12. **Tuple Sum**: Write a program that takes a tuple of integers as input and calculates the sum of all the elements in the tuple using a while loop. Return the sum.


```python
my_tuple = (10, 20, 30, 40)

i = 0
total = 0

while i < len(my_tuple):
  total += my_tuple[i]
  i += 1
print(total)
```
