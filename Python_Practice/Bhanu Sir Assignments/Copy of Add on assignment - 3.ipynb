13. OTP Generator

```python
import random

#method 1
def generate_otp(length=6):
  otp = random.randrange(100000, 999999)
  return otp
print("OTP:", generate_otp())

#Method 2

def otp_generator(n):
  otp = ""
  for i in range(n):
    otp += str(random.randint(0,9))
  return otp

print("OTP: ",otp_generator(6))
print("OTP: ",otp_generator(4))


#METHOD 3

def otp_generator(n):
  return "".join(random.choices('0123456789', k = n))

print("OTP: ",otp_generator(6))
print("OTP: ",otp_generator(4))
```

#Password Generator


```python

import string
import random

def password_generator(n):
  return "".join(random.choices(string.digits + string.ascii_letters + string.punctuation, k = n))

print("Password: ", password_generator(8))
print("Password: ", password_generator(12))
```
